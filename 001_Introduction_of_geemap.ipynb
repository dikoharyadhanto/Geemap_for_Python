## Add Earth Engine Python Script

```python
# Add Earth Engine dataset
import ee

Map = geemap.Map()
dem = ee.Image('USGS/SRTMGL1_003')
landcover = ee.Image("ESA/GLOBCOVER_L4_200901_200912_V2_3").select('landcover')
landsat7 = ee.Image('LE7_TOA_5YEAR/1999_2003').select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7'])
states = ee.FeatureCollection("TIGER/2018/States")

# Set visualization parameters.
vis_params = {
  'min': 0,
  'max': 4000,
  'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']}

# Add Earth Eninge layers to Map
Map.addLayer(dem, vis_params, 'SRTM DEM', True, 0.5)
Map.addLayer(landcover, {}, 'Land cover')
Map.addLayer(landsat7, {'bands': ['B3', 'B2', 'B1'], 'min': 20, 'max': 200, 'gamma': 2.0}, 'Landsat 7')
Map.addLayer(states, {}, "US States")
```

```python
for c in collections:
    print(c) #Check how many ID Images
```

```python
Map = geemap.Map()
nlcd_2001 = ee.Image('USGS/NLCD_RELEASES/2019_REL/NLCD/2001').select('landcover')
nlcd_2019 = ee.Image('USGS/NLCD_RELEASES/2019_REL/NLCD/2019').select('landcover')

left_layer = geemap.ee_tile_layer(nlcd_2001, {}, 'NLCD 2001')
right_layer = geemap.ee_tile_layer(nlcd_2019, {}, 'NLCD 2019')

Map = geemap.Map()
Map.split_map(left_layer, right_layer)
Map
```

# Use Drawing Tools

```python
import ee
# https://developers.google.com/earth-engine/datasets/catalog/USGS_SRTMGL1_003
# Add Earth Engine dataset
image = ee.Image("USGS/SRTMGL1_003")

#Set visualization parameters
vis_params = {
    'min' : 0,
    'max' : 4000,
    'palette' : ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']}

# Add Earth Engine DEM to Map
Map = geemap.Map()
Map.addLayer(image, vis_params, 'SRTM DEM')
Map
```

```python
Map = geemap.Map()
states = ee.FeatureCollection("FAO/GAUL/2015/level0")
Map.addLayer(image, vis_params, 'SRTM DEM')
Map.addLayer(states, {}, 'World States')

Map
```

```python
Map.draw_features # List of Draw Features
```

```python
# Total Draw Features
print('Total Draw Features: ', len(Map.draw_features))
```

```python
Map.draw_last_feature # Last Draw Feature
```

```python
roi = ee.FeatureCollection(Map.draw_features)
indonesia = states.filterBounds(roi)
Map.addLayer(indonesia, {}, 'Indonesia')
```

```python
clipped_image = image.clip(indonesia)
Map.addLayer(clipped_image, vis_params, 'Indonesia DEM')
```
