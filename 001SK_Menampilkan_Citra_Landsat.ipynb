```python
import ee
import geemap
```

#Convert JS Script Code to Python Code

```python
js_snippet = """
var dataset = ee.ImageCollection('LANDSAT/LC08/C01/T1') // buat nentuin image-nya. Kode image dicatet dulu di menu pencarian citra
                  .filterBounds(geometry) //fungsi untuk memfilter pencarian geomer atas sebuah dataset
                  .filterDate('2021-06-01', '2021-09-27'); // fungsi untuk pencarian berdasarkan waktu
print(dataset); //fungsi memunculkan dataset layer

var single = ee.Image('LANDSAT/LC08/C01/T1/LC08_122064_20210916') //buat nampilin image-nya
            
var RGBTrue = single.select(['B4', 'B3', 'B2']); //buat nampilin composite band True color

//memunculkan citra
var RGBparam = { min: 0, max: 20000,}; //mau nampilin citra dengan nilai pixel rentang berapa sampe berapa
Map.addLayer(RGBTrue, RGBparam, 'Nama_Citra-nya');

"""
     
```

```python
lines = geemap.js_snippet_to_py(js_snippet, add_new_cell=False, import_ee=True, import_geemap=True, show_map=True)
for line in lines:
    print(line.rstrip())
```

# Add Landsat Image

```python
dataset = ee.ImageCollection('LANDSAT/LC08/C01/T1') # buat nentuin image-nya. Kode image dicatet dulu di menu pencarian citra
dataset_filter = dataset.filterBounds(roi).filterDate('2021-06-01', '2021-09-27')
```

```python
data_info = dataset_filter.aggregate_array('system:id').getInfo()
print(len(data_info)) #fungsi memunculkan jumlah dataset layer dalam collection
```

```python
for c in data_info:
    print(c) #Check how many ID Images
```

```python
single = ee.Image('LANDSAT/LC08/C01/T1/LC08_122064_20210916') #buat nampilin image-nya
```
